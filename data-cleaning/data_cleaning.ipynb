```python
import emoji
import re

def process_emojis(text):
    if not isinstance(text, str) or text == '':
        return ""

    # 1. ÂÖàÂ∞Ü Emoji ËΩ¨Êç¢‰∏∫ÊñáÂ≠óÊ†áÁ≠æ (‰æãÂ¶Ç ‚ù§Ô∏è‚ù§Ô∏è ÂèòÊàê :red_heart::red_heart:)
    # Â¢ûÂä† delimiters Á°Æ‰øùÊ†áÁ≠æ‰πãÈó¥Ê≤°ÊúâÁ©∫Ê†ºÔºåÊñπ‰æøÊ≠£ÂàôÂåπÈÖç
    text = emoji.demojize(text, delimiters=(":", ":"))

    # 2. ÂÖ≥ÈîÆÊ†∏ÂøÉÔºöÂØπËøûÁª≠ÈáçÂ§çÁöÑ :Ê†áÁ≠æ: ËøõË°åÂéªÈáç
    # Ëøô‰∏™Ê≠£Âàô‰ºöÂåπÈÖçÂΩ¢Â¶Ç :any_name::any_name: ÁöÑÁªìÊûÑÂπ∂Âè™‰øùÁïô‰∏Ä‰∏™
    text = re.sub(r'(:[a-z0-9_-]+:)\1+', r'\1', text)

    # 3. Â§ÑÁêÜÊôÆÈÄöÁöÑÊ†áÁÇπÁ¨¶Âè∑ÈáçÂ§ç (Â¶Ç !!! ÂèòÊàê !)
    text = re.sub(r'([^\w\s:])\1+', r'\1', text)

    # 4. Ê†ºÂºèÁæéÂåñÔºöÂéªÊéâÂÜíÂè∑ÔºåÊää‰∏ãÂàíÁ∫øÊç¢ÊàêÁ©∫Ê†º
    text = text.replace(':', ' ').replace('_', ' ')
    
    # 5. ÊúÄÂêéÊ∏ÖÁêÜÔºöÂéªÊéâÂ§ö‰ΩôÁ©∫Ê†ºÔºåËΩ¨Â∞èÂÜô
    text = ' '.join(text.split()).lower()
    
    return text

# --- ÊµãËØïÊ°à‰æã ---
test_samples = [
    "So beautiful!!! ‚ù§Ô∏è‚ù§Ô∏è", 
    "Funny üòÇüòÇüòÇ", 
    "Good job üëèüëèüëè"
]

for s in test_samples:
    print(f"ÂéüÂßã: {s}  =>  ÁªìÊûú: {process_emojis(s)}")
```

```python
df_raw['text'] = df_raw['text'].apply(process_emojis)
```

```python
#
import html 
url_pattern = r'https?://\S+|www\.\S+'
df_raw['text'] = df_raw['text'].str.replace(url_pattern, '', regex=True)
df_raw['text'] = df_raw['text'].apply(html.unescape)
df_raw['text'] = df_raw['text'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
# Word Frequency Statistics:
from collections import Counter
all_words = ' '.join(df_raw['text'].astype(str).str.lower()).split()
word_counts = Counter(all_words)
df_word_freq = pd.DataFrame(word_counts.most_common(100), columns=['Word', 'Frequency'])
print("üìä ÂéüÂßãËØçÈ¢ëÁªüËÆ° Top 50ÔºàËßÇÂØüÂì™‰∫õÊòØÊ≤°Áî®ÁöÑÔºâÔºö")
with pd.option_context('display.max_rows', 50):
    display(df_word_freq.head(50))
```

```python
# build stop words removal list
stop_words_list = ['about','who','when','will',"it's",'the', 'and', 'to', 'this', 'i', 'of', 'you', 'with', 'it', 'in', 'is', 'for','john', 'christmas', 'that', 'advert', 'me', 'lewis', 'as', 'have', 'we', 'was', '-','my', 'your', 'ad', 'are', 'be', 'year', 'an', 'they', 'their', 'them', 'he', 'him', 'his', 'she', 'her', 'hers']
df_raw['text_final'] = df_raw['text'].apply(
    lambda x: ' '.join([
        word for word in str(x).lower().split() 
        if word not in stop_words_list and len(word) > 1
    ])
)
```
